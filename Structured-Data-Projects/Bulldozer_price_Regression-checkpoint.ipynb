## Building of Differnt Evaluatiion Metrics function


```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import root_mean_squared_log_error
np.random.seed(42)
evalution_method={'mean_absolute_error':mean_absolute_error,
                   'mean_squared_error':mean_squared_error,
                   'root_mean_squared_error':root_mean_squared_error,
                   'root_mean_squared_log_error':root_mean_squared_log_error
                 }
def evalution_metrics_func(y_true,y_preds,eval_methods):
    final_metrics={}
    for method,func in eval_methods.items(): 
        scores=func(y_true,y_preds)
        final_metrics[method]=scores

    return final_metrics



```

##  Hyperprameters tuning of model to get more ACCURACY...


```python
%%time
#lets try Random forest classifier model
from sklearn.ensemble import RandomForestRegressor 
from sklearn.model_selection import RandomizedSearchCV


rs_grid={
 'max_depth': [3,5,10, None],
 'max_features': [0.5,1,'auto', 'sqrt'],
 'min_samples_leaf': np.arange(1,20,2),
 'min_samples_split': np.arange(2, 20, 2),
 'n_estimators': np.arange(10,100,10),
  'max_samples':[100000]}

model=RandomForestRegressor(n_jobs=-1,
                           random_state=42)
rscv_model=RandomizedSearchCV(model,
                              rs_grid,
                              n_iter=10,
                              cv=10,
                              verbose=True)
#training of model
rscv_model.fit(x_train,y_train)
y_preds=rscv_model.predict(x_val[:10000])
```

```python
print(evalution_metrics_func(y_val[:10000],y_preds,evalution_method))
rscv_model.best_params_
```
